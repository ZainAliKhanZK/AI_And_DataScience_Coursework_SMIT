# **Hotel Booking Dataset**

# Importing libraries and Loading the Dataset

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
df = pd.read_csv("hotel_booking.csv")
```

# Identifying Null Values & Filling them out

```python
df.isnull().sum()
```

```python
df.country.unique()
```

```python
df.country.value_counts()
```

```python
df['country'] = df['country'].fillna('Unknown')
```

```python
df.children.value_counts()
```

```python
df['children'] = df['children'].fillna(df.children.mode()[0])
```

```python
df.agent.value_counts()
```

```python
df['agent'] = df['agent'].astype('Int64').astype(str).replace('<NA>', 'Direct')
```

```python
df['company'] = df['company'].fillna(0)
```

```python
df['reservation_status_date'] = pd.to_datetime(df['reservation_status_date'])
```

# Merging the arrival date column which was splitted in 3 columns

```python
df['arrival_date'] = pd.to_datetime(
    df['arrival_date_year'].astype(str) + '-' +
    df['arrival_date_month'] + '-' +
    df['arrival_date_day_of_month'].astype(str),
    format='%Y-%B-%d'
)
```

```python
df.head()
```

# Dropping the splitted Columns

```python
df = df.drop(columns=[
    'arrival_date_year',
    'arrival_date_month',
    'arrival_date_day_of_month',
    'arrival_date_week_number'
])
```

```python
df.head()
```

# Moving the arrival_date column

```python
col = df.pop('arrival_date')
df.insert(28, 'arrival_date', col)
```

# Visualizing 

# Cancellation rate by hotel type

```python
rate_hotel = df.groupby('hotel')['is_canceled'].mean() * 100
rate_hotel
```

```python
plt.figure(figsize=(6,4))
plt.title('Cancellation rate by hotel type')
plt.bar(rate_hotel.index,rate_hotel.values)
plt.show()
```

# Cancellation rate by cancellation history

```python
df['had_prev_cancel'] = df['previous_cancellations'] > 0
rate_prev = df.groupby('had_prev_cancel')['is_canceled'].mean() * 100

plt.figure(figsize=(6,4))
plt.bar(['No prior cancellations','Had prior cancellation'], rate_prev.values, color=['#2a78d6','#d64545'])
plt.title('Cancellation rate by cancellation history')
plt.ylabel('Cancellation rate (%)')
plt.show()
```
