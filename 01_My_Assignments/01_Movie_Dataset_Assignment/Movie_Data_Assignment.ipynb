### **Question 36**
(Visualization) 'Drama' aur 'Action' genres ki movies ke 'Gross' revenue ko compare karne ke liye ek Grouped Box Plot ya Violin Plot banayein.

```python
import matplotlib.pyplot as plt

drama = df[df['Genre'].str.contains('Drama') & df['Gross'].notna()]['Gross']
action = df[df['Genre'].str.contains('Action') & df['Gross'].notna()]['Gross']

plt.figure(figsize=(8, 6))
plt.boxplot([drama, action], labels=['Drama', 'Action'])
plt.yscale('log')  # because Gross values span from thousands to nearly $1B
plt.ylabel('Gross Revenue ($, log scale)')
plt.title('Gross Revenue: Drama vs Action')
plt.tight_layout()
plt.show()
```

### **Question 40**
(Visualization) Top 5 highest rated movies aur unke 'Meta_score' vs 'IMDB_Rating' (scaled to 100) ka ek Side-by-Side Comparison Bar Chart banayein.

```python
top5 = df.nlargest(5, 'IMDB_Rating')[['Series_Title', 'IMDB_Rating', 'Meta_score']]
```

```python
top5['IMDB_scaled'] = top5['IMDB_Rating'] * 10
```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(len(top5))  # positions for each movie
width = 0.35               # width of each bar

fig, ax = plt.subplots(figsize=(10, 6))
ax.bar(x - width/2, top5['IMDB_scaled'], width, label='IMDB (x10)', color='#2a78d6')
ax.bar(x + width/2, top5['Meta_score'], width, label='Meta Score', color='#eda100')

ax.set_xticks(x)
ax.set_xticklabels(top5['Series_Title'], rotation=30, ha='right')
ax.set_ylabel('Score (out of 100)')
ax.set_title('Top 5 Movies: IMDB Rating (scaled) vs Meta Score')
ax.legend()
plt.tight_layout()
plt.show()
```
