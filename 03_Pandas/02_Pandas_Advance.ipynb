```python
def calc_tax(salary):
    if salary >= 200000:
        return salary * 0.20
    elif salary >= 150000:
        return salary * 0.15
    elif salary >= 100000:
        return salary * 0.10
    elif salary >= 75000:
        return salary * 0.075
    elif salary >= 50000:
        return salary * 0.050
    elif salary >= 25000:
        return salary * 0.025
    elif salary >= 10000:
        return salary * 0.010
```

## APPLYING A FUNCTION TO A COLUMN

```python
df["Tax"] = df.Salary.apply(calc_tax) # OR df["Tax"] = df["Salary"].apply(calc_tax) 
```

```python
df["Tax2"] = df.Salary.apply(lambda sal: sal* 0.20 if sal >= 200000 
                            else sal * 0.15 if sal >= 150000 
                            else sal * 0.10 if sal >= 100000 
                            else sal * 0.075 if sal >= 75000 
                            else sal * 0.50 if sal >= 50000 
                            else sal * 0.010)
```
