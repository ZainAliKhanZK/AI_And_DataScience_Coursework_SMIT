```python
import yfinance as yf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import pandas_datareader as web
import plotly.express as px
import datetime
```

```python
year = 1
```

```python
end = datetime.datetime.today()
end
```

```python
# start = datetime.date(2025,07,12)
start = datetime.date(datetime.date.today().year - year, datetime.date.today().month, datetime.date.today().day)
```

```python
start
```

```python
sp500 = web.DataReader(["sp500"], 'fred', start, end)
sp500.tail()
```

```python
stocks_list = ["TSLA","AAPL","AMZN","GOOGL"]
```

```python
stocks_df = pd.DataFrame()

for stock in stocks_list:
    data = yf.download(stock, period = f'{year}y')
    stocks_df[stock] = data["Close"]
```

```python
stocks_df.head()
```

```python
stocks_df.reset_index(inplace=True)
```

```python
stocks_df
```

```python
sp500.reset_index(inplace=True)
```

```python
sp500
```

```python
sp500.rename(columns={"DATE" : "Date"},inplace=True)
```

```python
sp500
```

```python
data = pd.merge(stocks_df, sp500, on="Date", how="inner")
```

```python
data.set_index("Date",inplace=True)
```

```python
data
```

###  Functions to plot interactive plotly charts

```python
def interactive_plot(df):
        fig = px.line()
        for i in df.columns:
            fig.add_scatter(x=df.index, y=df[i], name=i)
        fig.update_layout(
            width = 900,
            height = 700,
            margin = dict(l = 20, r = 20, t = 50, b = 20),
            legend = dict(
                orientation = "h",
                yanchor = "bottom",
                y = 1.02,
                xanchor = "right",
                x = 1
            ) 
        )
        return fig
```

```python
interactive_plot(data)
```

```python
data_normalized = data/data.max()
```
