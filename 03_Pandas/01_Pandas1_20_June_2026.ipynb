```python
df['Tax'] = df.Salary*.10
df
```

```python
def calc_tax(Salary):
    if Salary>=200000:
        return Salary*.20
    elif Salary>=150000:
        return Salary*.15
    elif Salary>=100000:
        return Salary*.10
    elif Salary>=75000:
        return Salary*.075
    elif Salary>=50000:
        return Salary*.050
    else:
        return Salary*.010
```

```python
df['Tax'] = df.Salary.apply(calc_tax)
df
```

```python
df['Tax2'] = df.Salary.apply(lambda sal:
                            sal*0.20 if sal>=200000
                            else sal*.15 if sal>=150000
                            else sal*.10 if sal>=100000
                            else sal*.075 if sal>=75000
                            else sal*0.05 if sal>=50000
                            else sal*.010)
```
