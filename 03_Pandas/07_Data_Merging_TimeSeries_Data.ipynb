# Understanding Data Merging With Joins

|Types |  Description|
|------|-------------|
|INNER JOIN | Returns matching records from both tables.
|LEFT JOIN  | Returns all left table records and matched right table records, with nulls for missing matches.      |
|RIGHT JOIN | Returns all right table records and matched left table records, with nulls for missing matches.      |
|FULL JOIN  | Returns all records from both tables, matching where possible.                                       |
|CROSS JOIN | Produces a Cartesian product, matching every row from the first table with every row from the second.|
|SELF JOIN  | Joins a table to itself, treating it as two distinct tables.                                         |

# Inner means intersection (Merge if on both table)

```python
pd.merge(data0,data2, how= "inner",on="sales_man_id").drop(["sales_man_name_y"],axis=1)
```

# Outer means merge from both tables

```python
pd.merge(data0,data2, how= "outer",on="sales_man_id")
```

# Left means merge all from left table and matching from right table

```python
pd.merge(data0,data2, how= "left",on="sales_man_id")
```

# Right means merge all from left table and matching from right table

```python
pd.merge(data0,data2, how= "right",on="sales_man_id")
```
