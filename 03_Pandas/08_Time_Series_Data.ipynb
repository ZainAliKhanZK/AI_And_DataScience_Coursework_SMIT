```python
import yfinance as yf
import numpy as np
import pandas as pd
import pandas_datareader
import matplotlib.pyplot as plt
```

```python
apple = yf.download("AAPL", start="2025-07-10", end="2026-07-10")
google = yf.download("GOOGL", start="2025-07-10", end="2026-07-10")
```

```python
apple
```

```python
apple_close = apple["Close"]
google_close = google["Close"]
```

```python
apple_close
```

```python
plt.figure(figsize=(16,8))
plt.plot(apple_close, label="Apple")
plt.title("Apple & Google Stocks")
plt.grid()
plt.plot(google_close, label="Google")
plt.legend()
plt.show()
```

```python
apple_close_weekly = apple_close.resample("W").mean()
```

```python
apple_close_weekly.head()
```

```python
plt.figure(figsize=(12,6))
plt.grid()
plt.title("Weekly Chart Of Apple")
plt.plot(apple_close_weekly)
```

```python
stocks = ["AAPL","GOOGL","AMZN","MSFT","NVDA"]
```

```python
closing_price = pd.DataFrame()

for stock in stocks:
    closing_price[stock] = yf.download(stock,start="2025-07-10", end="2026-07-10")["Close"]
```

```python
closing_price
```

```python
plt.figure(figsize=(12,6))
plt.grid()
# plt.title("Weekly Chart Of Apple")
plt.plot(closing_price)
```
