## Drop the Missing Values

```python
df.dropna(axis=0) # DROP NULL ROWS
```

```python
df.shape[0]
```

```python
int(df.shape[1] * 0.90)
```

```python
df.dropna(axis=1 , how= "any")
```

```python
df.dropna(axis=0 , thresh=int(df.shape[1] * 0.90))
```

## FIll Missing Values

```python
df.Age.fillna(df.Age.mean())
```

```python
df.Department.mode()
```

```python
df.Department.mode()[0]
```

```python
df.Department.fillna(df.Department.mode()[0])
```

```python
df.Salary
```

```python
df.Salary.fillna(df.Salary.mean())
```

```python
df.fillna(
    {
    "Age" : df.Age.fillna(df.Age.mean()),
    "Department" : df.Department.mode()[0],
    "Salary" : df.Salary.mean()
    }
)
```
