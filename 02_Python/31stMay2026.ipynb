```python
class Person:
    # class attribute (class se, or instance)
    legs = 2    
    def __init__(self,gender,name,age,height,weight,):
        # instance attributes (instance will be mandatory to access instance attributes)
        self.gender = gender
        self.name = name
        self.age = age
        self.height = height
        self.weight = weight
        self.heartbeat = 72
    # instance methods 
    def sleep(self):
        print("Person is sleeping!")
    def eat(self):
        print("Person is eating!")
    def walk(self):        
        print("Person is walking!")
    
```

```python
class Teacher():
    def __init__(self,subject):
        self.subject = subject
    def teach(self):
        print("Teacher is teaching")
        
```

```python
class Doctor(Person,Teacher):
    #instance
    def __init__(self,gender,name,age,height,weight,degree,specializzation,experience):
        super().__init__(gender,name,age,height,weight)
        self.degree =degree
        self.specializzation =specializzation
        self.experience = experience
    def cure(self):
        print("Doctor is curing patients!")
    def prescribe_medicine(self):
        print("Doctor is prescribing medicine")
```

# Multilevel Inheritance

```python
# Define a class A with two functions
class A:
    def feature1(self):
        print("Feature1 is working")
    def feature2(self):
        print("Feature2 is working")
a1=A()
a1.feature1()
a1.feature2()
```

```python
# Define another class with two functions
class B():
    def feature3(self):
        print("Feature3 is working")
    def feature4(self):
        print("Feature4 is working")

b1 = B()
b1.feature3()
b1.feature4()
```

```python
class B(A):# Just by passing class name of parent class
    def feature3(self):
        print("Feature3 is working")
    def feature4(self):
        print("Feature4 is working")
# class was accessing only class B functions
# Now class B inherits class A, it can access all the
# attributes and behaviours of class A

b2 = B()

b2.feature3()
b2.feature4()
b2.feature1()
b2.feature2()
```

```python
# Define a class C that inherits class B directly
# and inherits class A indirectly
# Class C will have access to features of class A also

class C(B):
    def feature5(self):
        print("Feature5 is working")
    def feature6(self):
        print("Feature6 is working")
c1 = C()

c1.feature1() 
c1.feature2() 
c1.feature3() 
c1.feature4() 
c1.feature5() 
c1.feature6() 
```

# Multiple Inheritance

```python
# Define a class A with two functions
class A:
    def feature1(self):
        print("Feature1 is working")
    def feature2(self):
        print("Feature2 is working")
```

```python
class B():
    def feature3(self):
        print("Feature3 is working")
    def feature4(self):
        print("Feature4 is working")
```

```python
##########################
## Multiple Inheritacne###
##########################

class C(A,B):
    def feature5(self):
        print("Feature5 is working")
    def feature6(self):
        print("Feature6 is working")
# Now class C inherits class A and class B
# Now remember there is no relationship now between
# class A and class B drectly or indirectly
c11 = C()
c11.feature1()
c11.feature2()
c11.feature3()
c11.feature4()
c11.feature5()
c11.feature6()
```
