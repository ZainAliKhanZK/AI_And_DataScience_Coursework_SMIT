```python
class Car:
    def __init__(self,name,model,year,color,engine,seats):
        self.name = name
        self.model = model
        self.year = year
        self.color = color  #-----------------THESE ALL ARE ATTRIBUTES
        self.engine = engine
        self.seats = seats

    #------METHODS----

    def car_details(self):
        return f"{self.name} {self.model} {self.year} {self.color} {self.engine} {self.seats}"

    def start(self):
        print(f"{self.name} started!".title())

    def stop(self):
        print(f"{self.name} stopped".title())
                            #-----------------THESE ALL ARE FUNCTIONS
    def ac(self):
        print(f"{self.name} AC turned on".title())

    def exhaust(self):
        print(f"{self.name} exhaust on!".title())


car1 = Car("BMW","M3",1992,"Blue","1300 HP","4 Seater")
print(f"You have booked a testdrive for :\nBrand: {car1.name}\nModel: {car1.model}\nYear: {car1.year}\nColor: {car1.color}\nPower: {car1.engine}\nSeats: {car1.seats}")
```

```python
print(Car.car_details(car1))
```

```python
print(car1.car_details())
```
