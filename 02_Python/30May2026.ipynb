# Inheritance:
     

```python
class Car:    
    def __init__(self,make,model,year,color,seating, engine):# parameter
        # object variables/attributes
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.seating = seating
        self.engine = engine
        
   # methods
    def car_details(self):
        return f""" 
                    {self.make}
                    {self.model}
                    {self.year}
                    {self.color}
                    {self.seating}
                    {self.engine}
                    """
    def start(self):
        return f"{self.model}{self.make} is starting!!!!"
    def stop(self):
        return  f"{self.model}{self.make} is stopping!!!!"
    def fill_gas_tank(self):
        return f"{self.model}{self.make} is filling gass tank!!!!"
    def run(self):
        return f"{self.model}{self.make} is running fast!!!!"  
```

```python
class ElectricCar():
    def __init__(self,make,model,year,color,seating, engine):# parameter
        # object variables/attributes
        self.make = make
        self.model = model
        self.year = year
        self.color = color
        self.seating = seating
        self.engine = engine
        self.battery = "100amp"
        
    # methods
    def car_details(self):
        return f""" 
                    {self.make}
                    {self.model}
                    {self.year}
                    {self.color}
                    {self.seating}
                    {self.engine}
                    """
    def start(self):
        return f"{self.model}{self.make} is starting!!!!"
    def stop(self):
        return  f"{self.model}{self.make} is stopping!!!!"
    
    def run(self):
        return f"{self.model}{self.make} is running fast!!!!"  
```

```python
class ElectricCar(Car):
    battery = '40Amp'
```

```python
ec = ElectricCar('Tesla','City',2025,'white',5,'1300CC')
```

```python
print(ec.car_details())
```

```python
class ElectricCar(Car):
    def __init__ (self,make,model,year,color,seating, engine,owner,variant,mode,price):      
            
        super().__init__(make,model,year,color,seating, engine)
        
        self.owner =owner
        self.variant =variant
        self.mode = mode
        self.price = price    
```

```python
class Restaurant:
    def __init__(self,restaurant_name,cuisine_type):
        self.restaurant_name=restaurant_name
        self.cuisine_type =cuisine_type        
        self.number_served = 0
        
    def describe_restaurant(self):
        print(self.restaurant_name)
        print(self.cuisine_type)
        
    def open_restaurant(self):
        print(f"{self.restaurant_name} is open!")
    
    def set_number_served(self,new_number):
        self.number_served = new_number
        
```

```python
res = Restaurant('ABC','Chineese')
```

```python
res.restaurant_name
```

```python
res.cuisine_type
```

```python
res.describe_restaurant()
```

```python
res.open_restaurant()
```

```python
class IceCreamStand(Restaurant):
    flavours = ['mango','chocolate', 'vanilla','strawberry']
    
    def show_flavours(self):
        print("Available Flavours:")
        for f in flavours:
            print(f"-{f}")
```

```python
ics = IceCreamStand('MyIceCream','Gelato')
```
