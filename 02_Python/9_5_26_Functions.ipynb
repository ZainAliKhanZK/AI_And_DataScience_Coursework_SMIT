```python
def txt_analyser(txt):
    char_count = 0
    digit_count = 0
    special_count = 0

    for i in txt:
        if i.isalpha():
            char_count += 1
        elif i.isdigit():
            digit_count += 1
        elif i.isspace():
            continue
        else:
            special_count += 1

    print(f"Total Characters are: {char_count}\nTotal Digits are: {digit_count}\nTotal Special Characters are: {special_count}\n")

txt = "In Python, a string can contain different types of characters such as letters, digits, and special characters. Digits are numbers from 0 to 9 that are often used for counting or calculations. Special characters include symbols like @, #, !, and $ that are not letters or digits. When processing text, it is important to identify these different types so we can separate or handle them correctly in a program."

txt_analyser(txt)
```
