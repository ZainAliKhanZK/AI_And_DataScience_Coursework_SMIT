## #1 Encapsulation

- Binding data with Methods.

```python
class Student():
    def __init__(self,name,roll,course,email,marks):
        self.name = name
        self.roll = roll
        self.course = course
        self.email = email
        self.marks = marks

    def show_profile(self):
        print("-------Student Profile-------")
        print(f"{self.name}")
        print(f"{self.roll}")
        print(f"{self.course}")
        print(f"{self.email}")
        print(f"{self.marks}")

    def get_marks(self):
        return self.marks
    
    def set_marks(self,new_marks):
        if (new_marks>=0) & (new_marks<=100):
            self.marks = new_marks
        else:
            return "Invalid Marks"


s1 = Student("Abdullah","BAI-24F-020","BSAI","abdullah@gmail.com",550)
```

```python
print(s1.name)
```

```python
print(s1.show_profile())
```

```python
s1.set_marks(97)
```

```python
# Accessing a public attribute outside the class
s1.get_marks()
```

## Access Modifiers

- Public
- Private
- Protective


```python
class Student():
    def __init__(self,name,roll,course,email,marks):
        # Instance Variables
        # Public Access
        self.name = name
        self.roll = roll
        self.course = course
        self.email = email

        # Protected Instance Variable
        self._marks = marks

    def show_profile(self):
        print("-------Student Profile-------")
        print(f"{self.name}")
        print(f"{self.roll}")
        print(f"{self.course}")
        print(f"{self.email}")
        print(f"{self._marks}")

    def get_marks(self):
        return self.marks
    
    def set_marks(self,new_marks):
        if (new_marks>=0) & (new_marks<=100):
            self.marks = new_marks
        else:
            return "Invalid Marks"


s2 = Student("Ayub","BAI-24F-014","BSAI","abdullah@gmail.com",94)
```

```python
# Accessing a protected atttibute outside the class
s2._marks
```

```python
class Patient:
    def __init__(self,name,cnic,blood_group,Therapeutic_privilege):
        self.name = name
        self.cnic = cnic
        self.blood_group = blood_group 
        self._Therapeutic_privilege = Therapeutic_privilege

    def patient_summary(self):
        print(f"""---------------Patient's Report---------------
              \nPatient's Name: {self.name}
              \nPatient's CNIC: {self.cnic}
              \nPatient's Blood Group: {self.blood_group}
              \nPatient's Therapeutic Privilege: {self.__Therapeutic_privilege}""")
```

```python
patient1 = Patient("John",4230102293992,"O-Negative","Cancer")
```

```python
print(patient1.name)
```

```python
print(patient1.blood_group)
```

```python
patient1.patient_summary()
```
