```python
import random

random.randint(1000,9999)
```

```python
all_accounts = []
```

```python
def open_account(initial_deposit=0):
    account = {}

    account_title = input("Enter Your Full Name: ".title())
    f_name = input("Enter Your Father Name: ".title())
    contact_number = int(input("Enter Your Contact Number: "))
    
    cnic = input("Enter your CNIC: ")
    initial_deposit = int(input("Enter Initial Deposit: "))

    account_pin = random.randint(1000,9999)
    account_number = random.randint(10000,99999)

    account["Title"] = account_title
    account["Father's Name"] = f_name
    account["Contact Number"] = contact_number
    account["CNIC"] = cnic 
    account["Balance"] = initial_deposit
    
    account["Account Number"] = account_number
    account["Pin"] = account_pin

    all_accounts.append(account)

    # account_to_show = {}

    # for key,values in account.items():
    #     if key in ["Account Number","Pin"]:
    #         continue
    #     account_to_show[key] = values

    return f"Thank you for opening account with us\nDetails of you account:\n{account}".title()
    
```

```python
print(open_account())
```

```python
for i in all_accounts:
    print(i)
```

```python
def check_balance():
    account_num = int(input("enter your account number:".title()))
    pin = int(input("enter your account pin:".title()))

    for i in all_accounts:
        if account_num == i["Account Number"]:
             if pin == i["Pin"]:
                
                print(f"Current Balance: Rs.{i['Balance']:,}")
                break
             else:
                print("Incorrect Pin!")
                break
    else:
        print("Account Not Found!")
        
```

```python
check_balance()
```

```python
def deposit_amount():
        
        account_num = int(input("enter your account number:".title()))
        amount = int(input("Enter the amount to deposit:".title()))

        for account in all_accounts:
                if account_num == account["Account Number"]:
                        if amount > 0:
                            account["Balance"]+= amount
                            print(f"amount deposited successfully!.\nnew balance: Rs.{account['Balance']:,}".title())
                            break
                        else:
                            print("enter amount greater than 0!.".title())
        else:
            print("account not found!".title()) 
```

```python
deposit_amount()
```

```python
def withdraw_amount():
    account_num = int(input("please enter your account number:".title()))
    pin = int(input("enter pin:".title()))
    amount = int(input("withdraw amount:".title()))

    for account in all_accounts:
        if account_num == account["Account Number"]:
            if pin == account["Pin"]:
                if amount > 0: 
                    if amount <= account["Balance"]:
                        account["Balance"] = account["Balance"] - amount
                        print(f"you have withdrawn a amount of Rs.{amount:,}\nNew Balance: Rs.{account['Balance']:,}".title())
                        break
                    else:
                        print("Insufficient Balance.")
                        break                
                else:
                    print("please enter a valid amount.".title())          
            else:
                print("incorrect pin".title())
                break
    else:
        print("account not found!".title())
```

```python
withdraw_amount()
```
