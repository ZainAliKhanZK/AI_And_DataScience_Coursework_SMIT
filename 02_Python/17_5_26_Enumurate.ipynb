# Bank

```python
import random
all_accounts=[] # all the bank opened account will lie here
closed_accounts = [] # all the bank closed account will lie here
```

```python
def open_account(cnic,account_title, initial_deposit=0):
    account={'cnic':cnic, 'balance':initial_deposit,
             'account_number':None, 'account_title':account_title, 'pin':None}
    account['account_number'] = random.randint(1000000,9999999)
    account['pin'] = random.randint(1000,9999)
    all_accounts.append(account)
    return "Thank you for opening account with us!"
```

```python
def check_balance(account_number,pin):
    for acc in all_accounts:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                return acc['balance']
            else:
                return "Invalid Pin"
    else:
        return "Invalid Account Number"
```

```python
check_balance(1095154,3796)
```

```python
def withdrawl_cash(account_number, pin, amount):
    for acc in all_accounts:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                if acc['balance']>=amount:
                    acc['balance']-=amount
                    return f'You withdrawl Rs:{amount} Successfully!'
                else:
                    return 'insufficient balance'
            else:
                return "invalid pin"
    else:
        return "invalid ACCOUNT"
```

```python
def deposit_cash(account_number, amount):
    for acc in all_accounts:
        if acc['account_number']==account_number:
            acc['balance']+=amount
            return f"You deposited Rs:{amount} Succesfully!"
    else:
        return "invalid ACCOUNT"
```

```python
def transfer_amount(account_number, pin, amount,beneficiary_acc):
    for acc in all_accounts:
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                if acc['balance']>=amount:
                    for bacc in all_accounts:
                        if bacc['account_number']==beneficiary_acc:
                            acc['balance']-=amount
                            bacc['balance']+=amount
                            return f"You transfered Rs:{amount} to {bacc['account_title']} Successfully!"
                            
                    else:
                        return 'Invalid Beneficiary Account'
            else:
                return 'Invalid pin'
    else:
        return 'Invalid Account'
```

```python
def close_account(account_number, pin):
    for i,acc in enumerate(all_accounts):
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                closing_amt = acc['balance']
                closed_accounts.append(all_accounts.pop(i))
                return f"Collect your balance amount{closing_amt}. Your account closed successfully!"
            else:
                return 'invalid pin'
    else:
        return 'invalid accoumt'
```

```python
def change_pin(account_number, pin, newpin):
    for i,acc in enumerate(all_accounts):
        if acc['account_number']==account_number:
            if acc['pin']==pin:
                acc['pin'] == newpin
                return "Your pin changed successfully!"
            else:
                'invalid pin'
                
    else:
        return 'invalid accoumt'
```
